```python
#Arbitrary arguments (one * only)
def print_siblings(name, *siblings):
    print(name, "has the following siblings:")
    for sibling in siblings:
        print(sibling)
    print()
        
print_siblings("John", "Ashley", "Lauren", "Arthur", 'a', 'b', 'c')
print_siblings("Mike", "John")
print_siblings("Terry")
```

```python
#Keyword argument (two **)
def print_brothers_sisters(name, **sibs):
    print(name, "has the following siblings:")
    for i in sibs:
        print(i, ":", sibs[i])
    print()
    
print_brothers_sisters("John", Ashley="bro", Lauren="sis", Arthur="stepbro")
```
